It's irreducible because any state in the chain can be reached by any other state.

#### Activity 4

Compute the stationary distribution for the chain.

**Note:** The stationary distribution is a *left* eigenvector of the transition probability matrix associated to the eigenvalue 1. As such, you may find useful the numpy function `numpy.linalg.eig`. Also, recall that the stationary distribution is *a distribution*.

---

```python
import numpy as np
from scipy.linalg import eig 
import scipy
import numpy.linalg as linalg

transitionMatrix = np.array([[0, 1./6, 1./6, 1./6, 1./6, 1./6, 1./6, 0, 0],  ## GO
             [0, 0, 1./6, 1./6, 1./6, 1./6, 1./6, 1./6, 0],  ## VM
             [1./6, 0, 0, 1./6, 1./6, 1./6, 1./6, 1./6, 0],  ## JV
             [1./6, 1./6, 0, 0, 1./6, 1./6, 1./6, 1./6, 0],  ## VA
             [1./6, 1./6, 1./6, 0, 0, 1./6, 1./6, 1./6, 0],  ## FP
             [1./6, 1./6, 1./6, 1./6, 0, 0, 1./6, 1./6, 0],  ## MG
             [0, 0, 0, 0, 0, 0, 0, 0, 1],  ## GTJ
             [1./6, 1./6, 1./6, 1./6, 1./6, 1./6, 0, 0, 0],  ## PA
             [1./6, 0, 0, 1./6, 1./6, 1./6, 1./6, 1./6, 0]])  ## JAIL



M = transitionMatrix

#M = np.array([[2,2],
#              [1,3]
#             ])

values, vectors = np.linalg.eig(M.T)
    

print("\n Matrix \n")
print(M)

print("\n Values \n")
print(values.real)

print("\n Vectors \n")
print(vectors)

#print(vectors[:,1].real)

print("\n vector for value 1 \n")
print((M.T).dot(vectors[:,0]).real)

print("\nvalue that allows the vector be a distribution: ")
aux = 0
for x in (M.T).dot(vectors[:,0]).real:
    aux += x
    
y=1/aux

print(y)
#print(y*(M.T).dot(vectors[:,0]).real)

print("\nsum of all the values in the vector: ")
aux = 0
signal = 1
for i in y*(M.T).dot(vectors[:,0]).real:
    aux +=i 
    if i<0:
        signal = 0
print(aux)

print("\ncheck if all values in vector are positive: ")
if signal == 1:
    print("all values are positive")


print("\neigenVector for eigenValue 1 :")
sDistribution=[]
if np.allclose((M.T).dot(y*(M.T).dot(vectors[:,0]).real), 1*y*(M.T).dot(vectors[:,0]).real):
    sDistribution= (y*(M.T).dot(vectors[:,0]).real)
print sDistribution
```

#### Activity 5

Generate a 10,000-step long trajectory of the chain defined in Activity #1. 

---

```python
import numpy as np
import numpy.random as rnd 


count = 10000
position = 0
#pathNames = ["GO"]
pathStates = [0]
while count >0 :
    nextpos= (position + rnd.randint(1, 7)) % 8
    #pathNames+= [statesList[nextpos]]
    pathStates+= [nextpos]
   
    if nextpos == 6:
        nextpos= 8
        #pathNames+= [statesList[nextpos]]
        pathStates+= [nextpos]
        nextpos= 2
        
    position= nextpos
    count= count -1

#print pathNames
print pathStates
    
```
